## 1. Python is Slow<a class ='Python is Slow' id ='Topic1'>

<font size ="3">Scientific and numerical computing often requires processing massive datasets with complex algorithms. If you are a scientist or data professional, you want a programming language than can process data FAST. The closer a programming language is to machine instructions (binary), the faster it runs. That’s why for decades, programming languages like C, C++, and Fortran, were the to-go option for data-intensive applications in science and technology.
<br><br>
Python it is, indeed, easy to learn and write, but slow to run compared to compiled languages. Like really, really slow. For instance, computing the spectral-norm of a matrix, which is a common task in data applications, has been benchmarked at ~1.9 seconds in C, whereas python takes a whopping ~170.1 seconds, meaning that Python is ~90 times slower. Considering this, the fact Python became the dominant language in machine learning and data science is a bit of a puzzle.

There are at least two reasons for this why Python succeed anyways.

- The first is that as computing capacity became cheaper, processing time has become less important than coding-time. Basically, the time you save by writing code in high-level yet slow to-run programming languages compensates by their lack of performance at run time. And Python is exceptionally clean, intuitive, and easy to learn, compared to C or Fortran.
<br><br>
- The second reason, which is probably the major one, is libraries. Particularly libraries written in low-level high-performant languages. Turns out that Python extensibility allows programmers to write the “engine” for numerical computation in languages like C and Fortran, and then Python can invoke such “engines” in the background, meaning that although you write code in Python, it is executed in compiled C or Fortran code instead of Python itself. And that is how you obtain the best of both worlds: the easy and fast developing time of Python, plus the runtime performance of C or Fortran. True, there is some small overhead of going back and forth between languages, but its impact it’s minimal.

## 4. Numpy Arrays<a class ='Numpy Arrays' id ='Topic4'>

- <font size ="3">NumPy fundamental object is the *ndarray*. 
- Arrays are simply *ordered collections of elements*, like single numbers, lists, sets, vectors, matrices, or tensors. 
- In Additionally, *elements in an array are of the same data type*. For instance, an array can’t have integers and text at the same time. 
- The reason is simple: mathematical operations with objects containing multiple data types would be slow, and NumPy main goal is fast and efficient numerical computation.
- The “n” in “ndarray” makes references to the arbitrary number of dimensions it can take. 
- An array with one element and one dimension, it’s a “singleton” or just a number. 
- An array with four elements and two dimensions is a 2x2 matrix. 
- Put simply, an array is like an Excel sheet with the caveat that instead of being restricted to two dimensions, it can be extended to 3, 4, or higher dimensions, and that you can’t combine data types in a “sheet”.
- In NumPy, dimensions are called **axes**.

```python
#Importing the numpy
import numpy as np
```

- <font size ="3">**Creating an Array - array method**<br>
using "array method" constructor to build an array as:

```python
# 1 axis/dimensions array
one_dim=  np.array([1, 2, 3]) 

# 2 axis/dimensions array
two_dim_1=  np.array([[1, 2, 3]]) 

# 2 axis/dimensions array
two_dim_2= np.array([[1, 2, 3],
                     [4, 5, 6]])
 
# 3 axis/dimensions array
three_dim =  np.array([[[1, 2, 3], 
                        [4, 5, 6]],
                       [[1, 2, 3],
                        [4, 5, 6]]]) 
```

### b. Intrinsic NumPy array creation objects<a class ='Intrinsic NumPy array creation objects' id ='Section_5_2'>

<font size ="3">Manual input of data into arrays can be cumbersome, so NumPy offers a series of convenience methods to create arrays for special cases, like zeros, ones, and others. Below some common examples.

```python
# zeros
zeros = np.zeros(5)

# ones
ones = np.ones((3, 3))

# arange
arange = np.arange(1, 10, 2)

# empty
empty =  np.empty([2, 2])

# linspace
linespace = np.linspace(-1.0, 1.0, num=10)
linespace_epf = np.linspace(2.0, 3.0, num=5, endpoint=False)
linespace_retstep = np.linspace(2.0, 3.0, num=5, endpoint=False,retstep= True)
linespace_ept = np.linspace(2.0, 3.0, num=5, endpoint=True,retstep= True)

# full
full = np.full((3,3), -2)

# indices
indices =  np.indices((3,3))
```

```python
print(f'Array of zeros:\n{zeros}\n')
print(f'Array of ones:\n{ones}\n')
print(f'Array of empty entries:\n{empty}\n')
print(f'Evenly spaced array in a range:\n{linespace}\n')
print(f'Evenly spaced array in a range with endpoint as False:\n{linespace_epf}\n')
print(f'Evenly spaced array in a range with return step as True:\n{linespace_retstep}\n')
print(f'Evenly spaced array in a range: with return step and end point as True\n{linespace_ept}\n')
print(f'Array with same number on each entry:\n{full}\n')
print(f'Array from indices:\n{indices}\n')
```

- <font size ="3">The **zero method** generates an array of zeros of shape defined by a tuple passed to the function<br><br>
- The **ones method** generates an array of ones of shape defined by a tuple passed to the function<br><br>
- The **empty method** generates an empty array (although very small numbers will be printed) of shape defined by a tuple passed to the function<br><br>
- The **linespace method** generates an array of evenly spaced entries given a range and a step size<br><br>
- The **full method** returns an array of shape defined by a tuple passed to the function filled with the same value (third argument outside the tuple)<br><br>
- The **indices method** generates an array representing the indices of the grid

## 6. Vecotrization<a class ='Vectorization' id ='Topic6'>

- <font size ="3">“Pure” Python is slow. One of the culprits of such slowness is Python’s loops. Loops are bad for performance for complicated reasons related to Python design as a dynamically typed language. The shortest answer to why loops are slow is that Python takes multiple actions for each call (e.g., it access memory several times, type checking, etc.), that compound and hurt performance the more loops you execute.<br><br>
- In scientific computing we want speed, meaning we want to get rid of loops. This is precisely what’s vectorization all about: getting rid of loops by performing computations on multiple components of a vector at the same time. Hence, performing operations in “vector” form. In NumPy, vectors can be interpreted as an array, so we could call this **arrayization** 

### <font color ='red'>How vectorization looks like conceptually</font>


- <font size ="3">**Performance Check**<br>
let’s compare the performance gain of vectorization against looping in a simple sum.

```python
#generating randomg 100 numbers
x = np.random.rand(100)
y = np.random.rand(100)
```

```python
#Sum two array with a Python loop (non-vectorized)
%timeit for i in range(0, len(x)): x[i] + y[i] 
```

```python
#Sum arrays with NumPy (vectorized)
%timeit x+y
```

<font size ="3">As you can see, the *NumPy vectorized implementation* is several orders of magnitude faster.

## 7. Array mathematics and element-wise operations<a class ='Array mathematics and element-wise operations' id ='Topic7'>

### a. Array arithmetic<a class ='Array arithmetic' id ='Section_7_1'>

<font size ="3">As in regular mathematics, array arithmetic is fundamentally about addition, subtraction, multiplication, and division. In NumPy, this kind of operations are performed element-wise. 

In the above example, NumPy will add the first elements of each array together, the second elements of each array together, and the third elements of each array together.Hence, element-wise addition.

The same can be extrapolated to multi-dimensional arrays. Consider the following example:


The logic is the same: the top-left elements in each array are added together, the top-right elements of each array are added together, and so on. Subtraction, division, multiplication, exponentiation, logarithms, roots, and many other algebraic operations, will be performed in the same manner.


```python
#list of common arithmetic operations
a = np.arange(1, 10).reshape((3,3))
b = np.arange(10,19).reshape((3,3))

addition = a + b
subtraction = a - b
multiplication = a * b
true_division = a / b
floor_division = a // b
remainder = np.remainder(a, b) 
```

```python
print(f'Array a:\n{a}\n')
print(f'Array b:\n{b}\n')
print(f'Addition of a and b:\n{addition}\n')
print(f'Subtraction of a and b:\n{subtraction}\n')
print(f'Multiplication of a and b:\n{multiplication}\n')
print(f'True division of a and b:\n{true_division}\n')
print(f'Floor division of a and b:\n{floor_division}\n')
print(f'Remainder of a and b:\n{remainder}')
```

#### Let us see what will happen if we try to multiply a 3x3 array by a scalar
<font size ="3">1. The operation will fail, as their shapes do not match<br>
2. Just the first element of the array will be multiplied by the scalar<br>
3. All elements of the array will be multiplied by the scalar regardless<br>
<br>
Let us try this out

```python
array_scalar = a * 2
print(f'3x3 array:\n{a}\n')
print(f'3x3 array times an scalar:\n{array_scalar}')
```

<font size ="3">Each element of the array was multiplied by 2. How does this even work? 
- One option is to “loop” over each of array and multiply by 3 sequentially. But that it is slow, and NumPy is all about speed. <br><br>
- What happens is that the scalar is **“broadcast”** to match the shape of the array BEFORE multiplication. In practice, what we have is a 3x3 array times a 3x3 array of 2s as:


**Broadcasting will make computation way faster than looping.**

### d. Rounding<a class ='Rounding' id ='Section_7_4'>

<font size ="3">Rounding is a delicate subject as rounding errors when compounded over sequences of operations, can completely mess up your results. It is also a common operation for presenting and plotting results to others. Rounding is applied element-wise.

Let’s generate a sequence of random decimal numbers to see the effect of different rounding procedures

```python
decimals = np.linspace(0.11111111, 0.99999999, 10)

# rounding
around = np.around(decimals, 3)
# rounding
round_ = np.round(decimals, 3)
# rounding to integer
rint = np.rint(decimals)
# rounding integer towards zero
fix = np.fix(decimals)
# round to the floor
floor = np.floor(decimals)
# round to the ceiling
ceil = np.ceil(decimals)
```

```python
print(f"Array of decimals:\n{decimals}\n")
print(f"'around' round to the given number of decimals:\tnp.around(decimals, 3)\n{around}\n")
print(f"'round' yields identical results than 'around':\tnp.round(decimals, 3)\n{round_}\n")
print(f"'rint' round to the nearest integer:\tnp.rint(decimals)\n{rint}\n")
print(f"'fix' round to the nearest integer towars zero:\tnp.fix(decimals)\n{fix}\n")
print(f"'floor' round to the floor of the input:\tnp.floor(decimals)\n{floor}\n")
print(f"'ceil' round to the ceiling of the input:\tnp.ceil(decimals)\n{ceil}")
```

### e. Exponents and logarithms<a class ='Exponents and logarithms' id ='Section_7_5'>

<font size ="3">Exponents and logarithms are often used in computations related to probability and statistics.

```python
z = np.array([0.1, 1, np.e, np.pi])

# exponent
exp = np.exp(z)
# exponent(x) -1
expm1 = np.expm1(z)
# 2^P
exp2 = np.exp2(z)
# natural log
log = np.log(z)
# log base 10
log10 = np.log10(z)
# log base 2
log2 = np.log2(z)
```

```python
print(f'Compute exponential element-wise:\n{exp}\n')
print(f"Compute 'exp(x) - 1' with greater precision for small values:\n{expm1}\n")
print(f"Compute '2**p' for all elements p in the array:\n{exp2}\n")
print(f'Compute natural logarithm element-wise:\n{log}\n')
print(f'Compute base 10 logarithm element-wise:\n{log10}\n')
print(f'Compute base 2 logarithm element-wise:\n{log2}\n')
```

### f. Other miscellaneous element-wise operations<a class ='Other miscellaneous element-wise operations' id ='Section_7_6'>

<font size ="3">NumPy arrays can contain a single data type. This constraint makes data storing and manipulation much more efficient than working with mixed type arrays (like Python lists), which is a priority for NumPy.

Since Python is a **“dynamically typed”** language, which simply means that you do not have to tell the computer what data type you will use (Python does this for you), most of the time you do not need to indicate which data type you will use.

```python
bools = np.array([True, False])
ints = np.full((2,2), 1)
floats = np.ones(5)
unsigned = np.arange(3, dtype=np.uint8)
complexs = np.array([+1j, +2j, +3j])
unicode = np.array(["Catbug", "Chris", "Danny", "Wallow", "Beth"])
```

A final operation you may want to do, is to change the data type of an array. For instance, let’s say you want to stack two arrays, but one has a floating-point data type and the other integers. Recall that NumPy arrays can have a single data type. Let’s see an example.

This is simply because int64 can’t represent float point or real numbers, only integers or natural numbers. But float64 can represent integers. So it is the smart choice to make things work.

### d. Array broadcasting <a class='Array broadcasting' id ='Section_9_4'>

- <font size ="3">Broadcasting is an automatic NumPy mechanism to match the dimensionality of arrays with different shapes for element-wise operations. <br>
- Broadcasting usually improves speed by means of vectorizing operations, meaning that the loop will occur in compiled C code rather than in Python, as Python looping us is resource-intensive and slow. However, there are some cases where broadcasting is not the best option.

Till now we saw NumPy performs several important computations element-wise, which requires having arrays with matching shapes: arrays of shape (,1), (2,2), and (4,3,2), must be multiplied by arrays with shape (,1), (2,2), and (4,3,2), to be compatible. However, there are cases where we want to multiplied arrays with shapes that do not match, for instance:


Following linear algebra conventions, we should multiply each element of A by 2. The way to get around in this in NumPy, is by broadcasting the scalar to match the shape of A as:


The scalar only gets “stretched” vertically and horizontally during computation. Now, creating copies of x is memory inefficient, so NumPy does not actually copy the value in memory.
```python
array_one = np.arange(9)
array_two = np.arange(8).reshape((2,2,2))

split_one, split_two = np.split(array_one, 3), np.split(array_two, 2) 
array_split_one, array_split_two = np.array_split(array_one, 2), np.array_split(array_two, 3)
hsplit_one = np.hsplit(array_one, 3)
hsplit_two, vsplit_two = np.hsplit(array_two, 2), np.vsplit(array_two, 2)
```

```python
print(f'Array one (9,):\n{array_one}\nShape:{array_one.shape}\n')
print(f'Splitting Array one into 3 equal dimension:\tnp.split(array_one, 3)\n{np.split(array_one, 3)}\n')
print(f'Array one is split into 3 (1,3) sub-arrays:\n{split_one[0]}\n{split_one[1]}\n{split_one[2]}\n') 
print(f'Array two (2,2,2):\n{array_two}\nShape:{array_two.shape}\n')
print(f'Splitting Array two into 2 equal dimension:\tnp.split(array_two, 2)\n{np.split(array_two, 2)}\n')
print(f'Array two is split into 2 (1,2,2) sub-arrays:\n{split_two[0]}\n{split_two[1]}\n') 
```

<font size ="3">The array_split function provides identical functionality than the array function, with the difference that it will work even when the original array cannot be equally divided into the requested number of sub-arrays. Basically, if you try to split the array one (9,) into 2, the operation will work and the last number (the “9”) will be completely ignored. On the other hand, if you try to split an array two (2,2,2) into 3 parts, it will generate an extra empty axis.

```python
print(f'Array one (9,):\n{array_one}\nShape:{array_one.shape}\n')
print(f'Array one split into 2 sub-arrays:\t np.array_split(array_one, 2)\n{array_split_one[0]}\n{array_split_one[1]}\n')
print(f'Array two (2,2,2):\n{array_two}\nShape:{array_two.shape}\n')
print(f'Array two split into 3 sub-arrays:\tnp.array_split(array_two, 3)\n{array_split_two[0]}\n{array_split_two[1]}\n{array_split_two[2]}\n')
```

As with concatenation, horizontal split (hsplit) and vertical split (vsplit) provide equivalent functionality than the split method (split), but restricted to the horizontal and vertical axis respectively. Equal division is also a constrain here. Array one can’t be split vertically because it has only one dimension.

```python
print(f'Array one horizontal split into 3 sub-arrays:\tnp.hsplit(array_one, 3)\n{hsplit_one[0]}\n{hsplit_one[1]}\n{hsplit_one[2]}\n')
print(f'Array two horizontal split into 2 sub-arrays:\tnp.hsplit(array_two, 2)\n{hsplit_two[0]}\n{hsplit_two[1]}\n')
print(f'Array two horizontal split into 2 sub-arrays:\tnp.vsplit(array_two, 2)\n{vsplit_two[0]}\n{vsplit_two[1]}')
```

### g. Array repetition <a class='Array repetition' id ='Section_9_7'>

<font size ="3">Arrays can be constructed or expanded via repetition. **Data simulation** and **image manipulation** are situations where you may want to use this functionality.

There are two NumPy methods that may look similar at first but they are not: **tiling** and **repetition**.

```python
array_one = np.arange(1,4)
array_two = np.arange(1,10).reshape((3,3))

tile_one, tile_two, tile_three, tile_four = np.tile(array_one, (1,2)), np.tile(array_one, (2,1)), np.tile(array_two, (1,2)), np.tile(array_two, (2,1))  
repeat_one, repeat_two = np.repeat(array_one, 2, axis=0), np.repeat(array_two, 2) 
repeat_three, repeat_four = np.repeat(array_two, 2, axis=1), np.repeat(array_two, 2, axis=0)
```

<font size ="3">**Tiling** will attach an entire copy of the array (as a block) at its right or below it. The number of copies attached is specified as a tuple, with the first indicating the number of “rows” copies and the second the number of “column” copies.

```python
print(f'Array One:\n{array_one}\n')
print(f'Array Two:\n{array_two}\n')
print(f'Repeat array one twice column-wise:\tnp.tile(array_one, (1,2))\n{tile_one}\n')
print(f'Repeat array one twice row-wise:\t np.tile(array_one, (2,1))\n{tile_two}\n')
print(f'Repeat array two twice column-wise:\tnp.tile(array_two, (1,2))\n{tile_three}\n')
print(f'Repeat array two twice  row-wise:\tnp.tile(array_two, (2,1))\n{tile_four}\n')
```

The **repeat method** will replicate the elements of the array in place. For instance:<br>
&emsp;&emsp;&emsp;&emsp;&emsp;&emsp; **[1  2]**<br>
Will be repeated as:<br>
&emsp;&emsp;&emsp;&emsp;&emsp;&emsp; **[1  1  2  2]**<br>
Instead as:<br>
&emsp;&emsp;&emsp;&emsp;&emsp;&emsp; **[1  2  1  2]**<br>
The latter behavior is expected from the tile method instead. The axis for repetition is specified independently as “axis=0” for rows and “axis=1” for columns.

```python
print(f'Array One:\n{array_one}\n')
print(f'Array Two:\n{array_two}\n')
print(f'Repeat array one twice row-wise:\tnp.repeat(array_one, 2, axis=0)\n{repeat_one}\n')
print(f'Repeat array two twice and flattened into one dimension:\tnp.repeat(array_two, 2) \n{repeat_two}\n')
print(f'Repeat array two twice column-wise:\tnp.repeat(array_two, 2, axis=1)\n{repeat_three}\n')
print(f'Repeat array two twice row-wise:\tnp.repeat(array_two, 2, axis=0)\n{repeat_four}\n')
```

### i. Rearranging array elements <a class='Rearranging array elements' id ='Section_9_9'>

<font size ="3">By **rearranging** we refer to altering the order or position of the elements of an array without changing its shape

```python
array_one = np.arange(1,10)
array_two = np.arange(1,10).reshape((3,3))

flip_one, flip_two, flip_three, flip_four = np.flip(array_one), np.flip(array_two), np.flip(array_two, 0), np.flip(array_two, 1) 
roll_one, roll_two, roll_three, roll_four = np.roll(array_one, 1), np.roll(array_two, 1), np.roll(array_two, 1, axis=0), np.roll(array_two, 1, axis=1) 
```

<font size ="3">The **flip** reverse the order of elements in an array along the specified axis. 
- If no axis is specified, the order of the elements is reversed as if it were a flattened array, but the shape is preserved. 
- Notice that for arrays with 2 or more axis, flipping happens to entire rows or columns (or elements of the axis) rather than that to elements within rows or columns. <br>
There are many ways to flip an array by combining position and axis, here just a couple of examples.

```python
print(f'Array one:\n{array_one}\n')
print(f'Array two:\n{array_two}\n')

print(f'Reverse array one:\tnp.flip(array_one)\n{flip_one}\n')
print(f'Reverse array two:\tnp.flip(array_two)\n{flip_two}\n')
print(f'Reverse array two along axis 0 (rows):\tnp.flip(array_two, axis =0)\n{flip_three}\n')
print(f'Reverse array two along axis 1 (cols):\tnp.flip(array_two, axis=1)\n{flip_four} \n')
```

<font size ="3">The **roll method** moves or “push” elements in an array along the specified axis. 
- This has the effect of moving all elements at once, so all get repositioned. <br>There are many ways to roll an array by combining the number of positions to be roll and the axis, here just a couple of examples.

```python
print(f'Array one:\n{array_one}\n')
print(f'Array two:\n{array_two}\n')

print(f'Roll elements array one by one position:\tnp.roll(array_one, 1)\n{roll_one}\n')
print(f'Roll elements array two by one position:\tnp.roll(array_two, 1)\n{roll_two}\n')
print(f'Roll elements array two by one position along axis 0 (rows):\tnp.roll(array_two, 1, axis=0)\n{roll_three}\n')
print(f'Roll elements array two by one position along axis 1 (cols):\tnp.roll(array_two, 1, axis=1)\n{roll_four}')
```

## 11. Array Indexing<a class ='Array Indexing' id ='Topic11'>

- <font size ="3">It identify the position each element occupies in the array.
- Indexing in NumPy is 0-based, as in native Python, meaning that you start to count positions at 0 rather than at 1. Indexing is done by utilizing square brackets as ([]).
- Indexing is a versatile operation useful in a wide variety of cases. The most common ones are to insert (“assignment”) values, to extract (“reference”) values, to delete values, and to change values.

### a. Basic indexing in one-dimensional arrays<a class ='Basic indexing in one-dimensional arrays' id ='Section_11_1'>

NumPy support indexing in one and multiple dimensions. <br>Let’s explore a simple case with a couple of examples.

```python
array_one = np.arange(1,11)

print(f'Array one: {array_one}')
print(f'Array one dimensions: {array_one.ndim}, shape:{array_one.shape}')
```

<b>Elements in array are index with square brackets:

```python
print(f'Array one: {array_one}')
print(f'Select element at position [0]: {array_one[0]}')
print(f'Select element at position [5]: {array_one[5]}')
print(f'Select element at position [9]: {array_one[9]}')
print(f'Select element at position [-5]: {array_one[-5]}')
print(f'Select element at position [-1]: {array_one[-1]}')
```

<font size ="3">Notice that **array_one[9]** and **array_one[-1]** return the same value, which is the last element of the array. This shows that NumPy (as Python does) can index elements both ways: (1) from **left-to-right** starting at 0, and (2) from **right-to-left** starting at -1.


### b. Slicing one-dimensional arrays<a class ='Slicing one-dimensional arrays' id ='Section_11_2'>

<font size ="3">To select a range of elements, also known as slicing, we use the [:] notation:

```python
print(f'Array one: {array_one}')
print(f'Elements from position [0] to position [3]: {array_one[0:3]}')
print(f'Elements from position [5] to position [9]: {array_one[5:9]}')
print(f'Elements from position [-9] to position [-5]: {array_one[-9:-5]}')
print(f'Elements from position [-3] to position [-1]: {array_one[-3:-1]}')
print(f'Elements from position [3] to position [-1]: {array_one[3:-1]}')
```

<font size ="3">When we slice arrays to return sub-arrays, we can specify the stride, this is, how many steps we take when pointing at indices to retrieve array elements. The default stride is one.

```python
print(f'Array one: {array_one}')
print(f'Slice from position [0] to position [6] with stride [1]: {array_one[0:6:1]}')
print(f'Slice from position [0] to position [6] with stride [2]: {array_one[0:6:2]}')
print(f'Slice from position [-6] to position [-1] with stride [3]: {array_one[-6:-1:3]}')
```

### c. Basic indexing in multidimensional arrays<a class ='Basic indexing in multidimensional arrays' id ='Section_11_3'>

<font size ="3">Array indexing with multiple dimensions follows the same logic as with one dimension. The NumPy indexing model figure above illustrates this as well.

There are two ways to index arrays with multiple dimensions:

- referencing each dimension/axis within a single pair of square brackets with each dimension/axis separated by commas [,]
- referencing each dimension independently with as many square brackets as dimensions/axes [ ][ ]

```python
array_two = np.arange(1,10).reshape((3,3))
array_three = np.arange(1,9).reshape((2,2,2))

print(f'Array two dimensions/axes: \n{array_two}\n')
print(f'Array three dimensions/axes: \n{array_three}\n')
print(f'Array two dimensions: {array_two.ndim}, shape:{array_two.shape}')
print(f'Array three dimensions: {array_three.ndim}, shape:{array_three.shape}')
```

## 12. Array Iteration<a class ='Array Iteration' id ='Topic12'>

- <font size ="3">Iterating over arrays refers to the operation of “visiting” elements of an array in a systematic fashion. To iterate over array elements we utilize standard Python syntax **(for and while loops)** plus the functionality provided by the **nditer** iterator object.
- Though it is a performance hit, but there are circumstances where iteration is unavoidable.<br>
Let us see few examples

### a. Basic array iteration <a class ='Basic array iteration' id ='Section_12_1'>

```python
array = np.arange(1,10).reshape(3,3)

print(f"Two-dimensional array:\n{array}\n")
for element in np.nditer(array):
    print(element, end=' ')
```

```python
for element in array:
    print(element, end=' ')
```

<font size ="3">The **nditer object** iterates over the array matching the way on which the data is stored in memory. This means that regardless of how you “present” the array to the iterator, you will get back the elements in the same order. **This done simply because is faster.**

```python
print(f"Two-dimensional array:\n{array}\n")
print(f"Transposed (along main diagonal) two-dimensional array:\n{array.T}\n")

for element in np.nditer(array.T):
    print(element, end=' ')
```

If you were expecting to get back 1, 4, 7, 2, 5, 8, 3, 6, 9, you are not alone. I was expecting that order too. But that is not the order data is stored in memory. **To visit the elements in the order you would expect by looking at the array, you have to explicitly add you want the C order:**

```python
print(f"Two-dimensional array:\n{array}\n")
print(f"Transposed (along main diagonal) two-dimensional array:\n{array.T}\n")

for element in np.nditer(array.T.copy(order='C')):
    print(element, end=' ')
```

- <font size ="3">There may be times when you want to update the values of an array while iterating. 
- For instance, reinforcement learning models constantly update values after each iteration. 
- The default behavior of **nditer** is “read-only”, meaning it won’t let you change values. 
- Hence, you have to specify either “readwrite” or “writeonly” options to update values. 

- Additionally, you have to signal to nditer when you have finished iterating over values, as **nditer** needs to update the array with the new values. 
- This happens because nditer first save the updated values in a temporary memory space instead of changing values “on the fly”. 
- According to NumPy docs, there are two ways to do this:

    - using **nditer** as a context manager utilizing the Python with statement
    - calling the iterator’s **close method** at the end of the iteration process

```python
#np.nditer is readonly
array_to_update = np.arange(1,10).reshape(3,3)

print(f'Array to update:\n{array_to_update}\n')


for element in np.nditer(array_to_update):
    element[...] = element*2
print(f'Updated array (squared):\n{array_to_update}')
```

```python
array_to_update = np.arange(1,10).reshape(3,3)

print(f'Array to update:\n{array_to_update}\n')

with np.nditer(array_to_update,op_flags=['readwrite']) as iterator:
    for element in iterator:
        element[...] = element*2
print(f'Updated array (squared):\n{array_to_update}')
```

```python
#slow process
array_to_update = np.arange(1,10).reshape(3,3)

print(f'Array to update:\n{array_to_update}\n')


for element in array_to_update:
    element[...] = element*2
print(f'Updated array (squared):\n{array_to_update}')
```

<font size ="3">Although looping is not completely avoidable while iterating, there is a way to speed up iteration by partially vectorizing the innermost loop of the iteration. The details of how this work is not relevant from an applied perspective. What you want to know is that it will be faster and that you have to declare the **external_loop flag in the iterator**.<br> Let’s time both approaches:

```python
large_array = np.arange(10000)

%timeit for element in np.nditer(large_array):element*2
```

```python
large_array = np.arange(10000)

%timeit for element in np.nditer(large_array,flags =['external_loop']):element*2
```

### b. Broadcasting array iteration <a class ='Broadcasting array iteration' id ='Section_12_2'>

<font size ="3">If you find yourself having to iterate over multiple arrays with different shapes and dimensionality, the **nditer object** is smart enough to apply broadcast rules during iteration. For instance:

```python
one_dim_array = np.arange(3)

two_dim_array = np.arange(9).reshape(3,3)
print(f'One Dimension Array:\n{one_dim_array}\n')
print(f'Two Dimension Array:\n{two_dim_array}\n')
for x, y in np.nditer([one_dim_array, two_dim_array]):

    print(f'x: {x}, y:{y}')
```

<font size ="3">Here we can see that **nditer broadcast** the **one_dim_array** to match the **two_dim_array** so iteration does not break.

### c. Allocating outputs from iteration <a class ='Allocating outputs from iteration' id ='Section_12_3'>

<font size ="3">There are cases where you want to create a function utilizing the nditer object. This is, functions that take an array as input, iterate over the array elements, and instead of modifying the original array, they return the output somewhere else.

```python
def square(a):
    with np.nditer([a,None]) as it:
        for x,y in it:
            y[...] = x*x
        return it.operands[1]
```

```python
input_array = np.arange(1,6)
print(f'Input array: {input_array}')
print(f'Input array squared: {square(input_array)}')
```

## 13. Array shallow and deep copies <a class ='Array shallow and deep copies' id ='Topic13'>

To copy an array in NumPy can mean three different things:

- To put a new label
- To create a “view” or “shallow copy” that refers to the same chunk of data in memory
- To create an independent or “deep copy” of the array in a different location in memory
<br><br>
Creating “shallow copies” instead of “deep copies” can significantly speed up computation and save space, but it has limitations to keep in mind. The figure below illustrates the differences between the three alternatives.



### a. Array new label<a class ='Array new label' id='Section_13_1'>

<font size ="3">There are cases where no copy at all is created. All that is done is attaching a new label to the original array. We can use the **is** and **may_share_memory** methods to check object identity and memory sharing.

```python
array_one = np.ones(10)
array_two = array_one

print(f'Array one:{array_one}\n')
print(f'Array two:{array_two}\n')
print(f'Are array-one and array-two the same object?: {array_two is array_one}\n')
print(f'Do array-one and array-two share memory?: {np.may_share_memory(array_one, array_two)}\n')
print(f'Identifier of Array one:{id(array_one)}')
print(f'Identifier of Array two:{id(array_two)}\n')
```

<font size ="3">The key here is that array_one and array_two share memory AND are the same object.

A logical consequence of the fact that **array_one and array_two share the same data and identity**, is that if you change array_two you will be inadvertently changing array_one as well:

```python
array_two[-1] = 99
print(f'New array-two:\n{array_two}\n')

print(f'Array-one is changed although no explicit operation was done to it:\n{array_one}')
```

<font size ="3">If you want two independent copies of the same data, you need a “deep copy” 

### b. Array shallow copy or view<a class ='Array shallow copy or view' id='Section_13_2'>

<font size ="3">“Shallow copies” or “views” are objects which are not the same, but share the same data source.

```python
array_three = array_one[0:6]

print(f'Array one:{array_one}\n')
print(f'Array three:{array_three}\n')
print(f'Are array-one and array-three the same object?: {array_two is array_three}\n')
print(f'Do array-one and array-three share memory?: {np.may_share_memory(array_one, array_three)}\n')
print(f'Identifier of Array one:{id(array_one)}')
print(f'Identifier of Array three:{id(array_three)}\n')
```

<font size ="3">**The key here is that array_one and array_three are NOT the same object BUT share memory.**

The effect of shallow copies in the base-data is a bit trickier. There are cases where changing a view does not change the base-data, as reshaping:

```python
array_three_reshape = array_three.reshape(2,3)

print(f'Array-three change shape:\n{array_three_reshape}\n')
print(f'But this does not change array-two (the source):\n{array_two}')
```

<font size ="3">**And cases where it does change the original base-data, as inserting new values:**

```python
array_three[0] = 99

print(f'Array-three new value at position 0:\n{array_three}\n')
print(f'This does change array-two (the source):\n{array_two}\n')

print(f'Are array-two and array-three the same object?: {array_two is array_three}')
print(f'Do array-two and array-three share memory?: {np.may_share_memory(array_two, array_three)}\n')

print(f'Identifier of Array one:{id(array_two)}')
print(f'Identifier of Array three:{id(array_three)}')
```

### c. Array deep copy<a class ='Array deep copy' id='Section_13_3'>

<font size ="3">“Deep copies” are independent copies located in a different position in memory.

```python
array_four = array_one[0:5].copy()

print(f'Array one:{array_one}\n')
print(f'Array four:{array_four}\n')
print(f'Are array-one and array-four the same object?: {array_one is array_four}')
print(f'Do array-one and array-four share memory?: {np.may_share_memory(array_one, array_four)}\n')
print(f'Identifier of Array one:{id(array_one)}')
print(f'Identifier of Array three:{id(array_four)}\n')
```

```python
array_four[3] = 99

print(f'Array-four new value at position 3:\n{array_four}\n')
print(f'This does NOT change array-one (the source):\n{array_one}')
```

<font size ="3">**Deep copies are recommended when you want to keep a subset of the data and throw away the base array, or when you need to manipulate two or more copies of the same data independently.**

### d. Setting a seed for reproducibility <a class ='Setting a seed for reproducibility' id ='Section_15_4'>

<font size ="3">In the previous examples, there is no way to reproduce the numbers we generated because we didn’t specify a “seed”. “Wait, we are generating random numbers, why do I would want to repeat the same random number?”. To allow others to reproduce your results. In Data Science and Machine Learning different starting points (seeds) may lead to widely different results.

<font size ="3">**Sampling from a binomial distribution example:**

<font size ="3">**binomial(n, p, size=None)**
<br>
Draw samples from a binomial distribution.
<br>
Samples are drawn from a binomial distribution with specified
parameters, n trials and p probability of success where
n an integer >= 0 and p is in the interval [0,1]. (n may be
input as a float, but it is truncated to an integer in use)

The probability density for the binomial distribution is

$P(N) = \binom{n}{N}p^N(1-p)^{n-N},$

```python
n, p1, p2, t = 1, .5, .8, 1000
binomial_fair = rg.binomial(n, p1, t)
binomial_bias = rg.binomial(n, p2, t)
```

<font size ="3">**Sampling from a chisquare distribution example:**

<font size ="3">**chisquare(df, size=None)**
<br>
Draw samples from a chi-square distribution.
<br>
When `df` independent random variables, each with standard normal
distributions (mean 0, variance 1), are squared and summed, the
resulting distribution is chi-square (see Notes).  This distribution
is often used in hypothesis testing.

The variable obtained by summing the squares of `df` independent,
standard normally distributed random variables:

$Q = \sum_{i=0}^{\mathtt{df}} X^2_i$

is chi-square distributed, denoted

$Q \sim \chi^2_k.$

The probability density function of the chi-squared distribution is

$p(x) = \frac{(1/2)^{k/2}}{\Gamma(k/2)}
                 x^{k/2 - 1} e^{-x/2},$

where $\Gamma$ is the gamma function,

$\Gamma(x) = \int_0^{-\infty} t^{x - 1} e^{-t} dt.$

```python
chisquare1 = rg.chisquare(5,1000)
chisquare2 = rg.chisquare(50,1000)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, constrained_layout=True)
fig.suptitle('Sampling from chisquare distribution')
ax1.hist(chisquare1, bins=50)
ax1.set_title("5 degrees of freedom")
ax2.hist(chisquare2, bins=50)
ax2.set_title("50 degrees of freedom");
```

<font size ="3">**poisson(lam=1.0, size=None)**
<br>
Draw samples from a Poisson distribution.
<br>
The Poisson distribution is the limit of the binomial distribution
for large N.
<br>
**The Poisson distribution**

$f(k; \lambda)=\frac{\lambda^k e^{-\lambda}}{k!}$

For events with an expected separation :math:$\lambda$ the Poisson
distribution $f(k; \lambda)$ describes the probability of
:math:`k` events occurring within the observed
interval $\lambda$.

Because the output is limited to the range of the C int64 type, a
ValueError is raised when `lam` is within 10 sigma of the maximum
representable value.


```python
poisson1 = rg.poisson(5, 1000)
poisson2 = rg.poisson(50, 1000)
```

### d. Singular value decomposition <a class ='Singular value decomposition' id ='Section_17_4'>

<font size ="3">The Singular Value Decomposition (SVD) of a matrix is a factorization of that matrix into three matrices.
Mathematics behind SVD
The SVD of  mxn matrix A is given by the formula :

$A = UWV^{T}$

where:

U:  mxn matrix of the orthonormal eigenvectors of $AA^{T}$.<br>
$V^{T}$: transpose of a nxn matrix containing the orthonormal eigenvectors of $A^{T}A$.<br>
W:  a nxn diagonal matrix of the singular values which are the square roots of the eigenvalues of $A^{T}A$.<br>


```python
U, S, T = np.linalg.svd(C)

print(f'Left orthogonal matrix C:\n{np.round(U, 2)}\n')
print(f'Singular values diagonal matrix C:\n{np.round(S, 2)}\n')
print(f'Right orthogonal matrix C:\n{np.round(T, 2)}')
```

## 19. Sorting Arrays<a class ='Sorting Arrays' id ='Topic19'>

### a. Sorting <a class ='sorting' id ='Section_19_1'> 

- #### Sort Method

```python
array_1 = np.array([2, 1, 5, 3, 7, 4, 6, 8])
array_2 = np.array([[1,4],[3,1]])

#sorting1-D Array
sort_1d = np.sort(array_1)

# sort along the last axis
sort_2d_la = np.sort(array_2)

# sort the flattened array
flattened_array = np.sort(array_2, axis=None)  

# sort along the first axis
sort_2d_fa = np.sort(array_2, axis=0) 

#Use the `order` keyword to specify a field to use when sorting a structured array
dtype = [('name', 'S10'), ('height', float), ('age', int)]
values = [('Arthur', 1.8, 41), ('Lancelot', 1.9, 38),('Galahad', 1.7, 38),('James', 1.8, 40)]
# create a structured array
structured_array = np.array(values, dtype=dtype) 

#sort by height
structured_array_sort_h = np.sort(structured_array, order='height') 

#Sort by age, then height if ages are equal:
structured_array_sort_ah = np.sort(structured_array, order=['age', 'height']) 
```

- #### Lexsort Method

<font size ="3">Given multiple sorting keys, which can be interpreted as columns in a
spreadsheet, lexsort returns an array of integer indices that describes
the sort order by multiple columns. 

The last key in the sequence is used
for the primary sort order, the second-to-last key for the secondary sort
order, and so on. The keys argument must be a sequence of objects that
can be converted to arrays of the same shape. If a 2D array is provided
for the keys argument, its rows are interpreted as the sorting keys and
sorting is according to the last row, second last row etc.

```python
first_names = np.array(['Betsey', 'Shelley', 'Lanell', 'Betsey', 'Margery'])
last_names = np.array(['Battle', 'Brien', 'Plotner', 'Stahl', 'Woolum'])
indx = np.lexsort((last_names, first_names)) 
```

```python
print(f'Array:\n{[first_names[i] + " " + last_names[i] for i in range(len(first_names))]}\n')
print(f'sorted index positions:\t{indx}\n')
print(f'Sorted Array:\n{[first_names[i] + " " + last_names[i] for i in indx]}')
```

- #### Maximum Method

- #### Minimum Method

```python
array_1 = np.array([5,3,7,12,57])
array_2 = np.array([8,1,5,31,34])

new_array = np.minimum(array_1,array_2)
print(f'Array 1:\n{array_1}\n')
print(f'Array 2:\n{array_2}\n')
print(f'Element wise Comparision between Array-1 and Array-2, returns minimum element :\n{new_array}\n')
```
