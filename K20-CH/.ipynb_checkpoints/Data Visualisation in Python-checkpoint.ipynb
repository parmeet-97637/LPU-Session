    - You could perform transformatioins in the column which has outlier

# Feature Engineering

### 1. Age : we will create the Age-Bands
we are converting the continuous feature to discrete feature

Min age :0.42
Max age : 80 
Age range : 0-80 and I want to create 5 bins, 80/5 = 16 so my bin size would 16
    
0-16  = Band-0
16-32 = Band-1
32-48 = Band-2
48-64 = band-3
>64   = Band-4

```python
df['Age_band']=0
df.loc[df['Age']<=16,'Age_band'] = 0 
df.loc[(df['Age']>16) & (df['Age']<=32),'Age_band'] = 1 
df.loc[(df['Age']>32) & (df['Age']<=48),'Age_band'] = 2 
df.loc[(df['Age']>48) & (df['Age']<=64),'Age_band'] = 3 
df.loc[(df['Age']>64),'Age_band'] = 4
```

```python
df.head()
```

```python
df['Age_band'].value_counts().to_frame().style.background_gradient(cmap='summer_r')
```

```python
sns.factorplot('Age_band','Survived',data=df,col='Pclass')
```

## Fare
fare_range

```python
df['Fare_range'] = pd.qcut(df['Fare'],4)
```

```python
df.head()
```

```python
df.groupby('Fare_range')['Survived'].mean().to_frame().style.background_gradient(cmap='summer_r')
```

- As the fare increases the survival chances increases.

```python
# converting the fare-range to bands
df['Fare_band'] = 0
df.loc[df['Fare']<=7.91,'Fare_band']=0
df.loc[(df['Fare']>7.91) & (df['Fare']<=14.454),'Fare_band'] =1
df.loc[(df['Fare']>14.454) & (df['Fare']<=31.0),'Fare_band'] =2
df.loc[(df['Fare']>31.0) & (df['Fare']<=513.0),'Fare_band'] =3
```

```python
sns.factorplot('Fare_band','Survived',hue='Sex',data=df)
```

## Convert the string values to numeric

```python
df['Sex'].replace(['male','female'],[0,1],inplace=True)
df['Embarked'].replace(['S','Q','C'],[0,1,2],inplace=True)
df['Initials'].replace(['Mr','Mrs','Miss','Master','Other'],[0,1,2,3,4],inplace=True)
```
