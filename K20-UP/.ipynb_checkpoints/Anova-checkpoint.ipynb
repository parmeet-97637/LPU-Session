### one-way Anova

<font size="3">
It is type of hypothesis test where only one factor is considered. We use F-statistics perform one-way Anova.

<font size="3">
    
**Steps Involved**

```python
Salary : High, Low, Medium : 100
        100-3 =97
    

Sex : Male and Female
    
k-1 : 2-1 = 1
    

```

1. Define the Null and the Alternate Hypothesis<br>
    H0(Null Hypothesis) : u1=u2=u3 (where u =mean)<br>
    H1(Alternate Hypothesis) : At least one difference amongst the means   <br> 
<br>
2. Finding the degree of fredom between the groups and within the groups<br>
    $df_{between} = k-1$<br>
where k is the number of groups

    $df_{within} = n-k$ <br>
where k is the number of groups<br>
n = number of samples in all groups combined<br>
<br>
        $df_{total} = df_{between} + df_{within}$
        
<br><br>

**F-Value:** It is defined as the ratio of variance between the samples to variance within the samples.
        $F-Value = \frac{Variance_{between-samples}}{Variance_{within-samples}}$

<br><br>
3. Refering the F-distribution table <a href = "http://www.socr.ucla.edu/Applets.dir/F_Table.html">F-Table</a> and find $F_{table}$ using $df_{between}$ and $df_{within}$<br>
$df_{1} = df_{between}$<br>
$df_{2} = df_{within}$

<br><br>

4. Find the mean of all samples in each group<br>
    $\mu_{Grand} = \frac{\sum{G}}{n}$
    
    where G = sum of all sample values<br>
    n = number of samples

5. Find the sum of squares of the total and sum of squares within.<br>
    $ SS_{total} = \sum{(X_{i} - \mu_{Grand})^2} $
    <br><br>
    $ SS_{within} = \sum{(X_{i} - \mu_{i})^2} $
    
    where $\mu_{i}$ is the means of the ith group
    <br>
    
    $SS_{between} = SS_{total} - SS_{within}$
    <br>
6. Find the variance between and within the groups<br>
    $S_{between}^2 = \frac{SS_{between}}{df_{between}} $ 
    
    $S_{within}^2 = \frac{SS_{within}}{df_{within}} $ 
    <br>
7. find Fcalc<br>
    $ F_{calc} = \frac{S_{between}^2}{S_{within}^2}$

## Interpret the results
if F_calc < F_table :
    Don't reject null hypothesis
    all the means are equal

if F_calc >F_table :
    Reject the null hypothesis

## Example

The marks of 3 subjects(out of 5) for a group of studet is recorded.
Signficance level : 5%

```python
import pandas as pd

cols = ['English(e)','Maths(m)','Science(s)']
marks = [[2,2,1],[4,3,2],[2,4,5]]

pd.DataFrame(marks, columns=cols)
```

Null Hypothesis :  $\mu_{e} = \mu_{m} = \mu_{s}$ <br>
Alternate Hypothesis : At least one difference among the mean of the 3 subjects

### degree of freedom 
k = 3
n = 9<br>
<br>
$df_{between} = k-1 = 3-1 = 2$<br>
$df_{within}  = n-k = 9-3 = 6$<br>
$df_{total}   = df_{between} + df_{within} = 2 + 6 = 8 $

### Step-3
df1 = 2 and df2 = 6 and alpha(significance level = 0.05) and for this valuces checking F_table= 5.1433

### Step-4

$\mu_{e} = (2+4+2)/3 = 2.67 $<br>
$\mu_{m} = (3+4+2)/3 = 3 $<br>
$\mu_{s} = (1+5+2)/3 = 2.67 $<br>
<br>
$\mu_{grand} = (8 + 9 + 8)/9 = 25/9 = 2.78$


### Step-5

$SS_{total} =  (2-2.78)^2 +(4-2.78)^2 +(2-2.78)^2 + (2-2.78)^2 +(3-2.78)^2+(4-2.78)^2 +(1-2.78)^2 +(2-2.78)^2 + (5-2.78)^2 = 13.60 $

$SS_{within} = (2-2.67)^2 +(4-2.67)^2 + (2-2.67)^2 + (2-3)^2+(3-3)^2+(4-3)^2 +(1-2.67)^2+(2-2.67)^2+(5-2.67)^2 = 13.34$

$SS_{between} = SS_{total} = SS_{within} = 13.60 - 13.34 = .26 $

## Step-6

$S_{between}^2 = .26/2 = .13$<br>
$S_{within}^2 = 13.34/6 = 2.22$


## Step-7

$F_{calc} = 2.22/.13 = .05 $

```python
F_calc = 0.5
F_table = 5.1433
F_calc<F_table
```

Conclusion : Don't reject null hypothesis all the means are equal

Thus, we can say that the means of all the subjects are same

## Two-way Anova in Python

Scientists needs to know whether the plant growth is affected by fertilizers and water frequency. 
They planted 30 plants and allowed them to grow for six months under different conditions for fertilizers and water frequency. 
After exactly siz months they recorded the height of the plants in cm. Perform the two-way Anova using pythong

```python
import pandas as pd
import numpy as np
```

```python
# create a dataframe
dataframe = pd.DataFrame({'Fertilizers' : np.repeat(['daily','weekly'],15), 
                          'Watering' : np.repeat(['daily','weekly'],15),
                         'height' : [14,16,15,15,16,13,12,11,14,
                                     15,16,16,17,18,14,13,14,14,
                                     14,15,16,16,17,18,14,13,14,
                                     14,14,15]})
```

```python
dataframe
```

```python
## conducting two-way Anova test
### To perfom two-way anova test we would be using statsmodel library
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
# perfomring two way-anova

model = ols('height ~ C(Fertilizers) + C(Watering) + \
    C(Fertilizers):C(Watering)' , data =dataframe).fit()
```

```python
sm.stats.anova_lm(model, type=2)
```

# Interpret the Results
- The fertilizers p-value is equal to .913305
- The Watering p-value is equal to .0546828
- The Fertilizer * Watering p-value is equal to .827527

Fertilizers & Watering is not having a statistical relationship in determining the height as the p-value .827 is grater than .05
