## Imputation

```python
df.loc[(df['Age'].isnull()) & (df.Initals=="Master"),"Age"] = 5
df.loc[(df['Age'].isnull()) & (df.Initals=="Miss"),"Age"] = 22
df.loc[(df['Age'].isnull()) & (df.Initals=="Mr"),"Age"] = 33
df.loc[(df['Age'].isnull()) & (df.Initals=="Mrs"),"Age"] = 36
df.loc[(df['Age'].isnull()) & (df.Initals=="Other"),"Age"] = 46
```

# Correlation between features

$ corr = \frac{\sum (x_i - \bar{x})(y_i - \bar{y})}{\sqrt(\sum (x_i -\bar{x})^2 \sum(y_i - \bar{y})^2)}$

```python
sns.heatmap(df.corr(),annot=True, cmap='RdYlGn',linewidth=0.2)
fig = plt.gcf()
fig.set_size_inches(10,8)
plt.show()
```

# Feature Engineering & Data Cleaning

Age Feature and convert it to Bands. ( Converting this contrinuous feature to discrete features)

- Bins size =  16
- Max passenger = 80
- 5 bins 

```python
# initlaizing a new columns
df['Age_Band'] = 0

df.loc[df['Age']<=16,'Age_Band'] = 0

df.loc[(df['Age']>16) & (df['Age']<=32),'Age_Band'] =1

df.loc[(df['Age']>32) & (df['Age']<=48),'Age_Band'] =2

df.loc[(df['Age']>48) & (df['Age']<=64),'Age_Band'] =3

df.loc[(df['Age']>64),'Age_Band'] =4
```

```python
df.head()
```

```python
df['Age_Band'].value_counts().to_frame().style.background_gradient(cmap='summer_r')
```

```python
sns.factorplot('Age_Band','Survived',data=df, col='Pclass')
```

- As your age increases the survival ratio decreses.

```python
## Parch +sibSp -> Family_size
df['Family_size'] = 0

df['Family_size'] = df['Parch'] + df['SibSp']

df['Alone'] = 0 
df.loc[df['Family_size'] ==0,"Alone"] = 1

df.head()
```

```python
f,ax = plt.subplots(1,2,figsize=(10,4))
sns.factorplot('Family_size','Survived',data=df,ax=ax[0])
ax[0].set_title("Family_size Vs Survived")

sns.factorplot('Alone','Survived',data=df,ax=ax[1])
ax[1].set_title("Alone Vs Survived")

plt.show()
```

```python
sns.factorplot("Alone","Survived", data=df,hue="Sex",col="Pclass")
```

# Fare

```python
# Range of fare
df['Fare_Range'] = pd.qcut(df['Fare'],4)
df.groupby('Fare_Range')['Survived'].mean().to_frame().style.background_gradient(cmap='summer_r')
```

- As the fare range increases the survival chances increases

```python
df['Fare_Band'] = 0

df.loc[df['Fare'] <=7.91,'Fare_Band'] = 0
df.loc[(df['Fare']>7.91) & (df['Fare']<=14.454),'Fare_Band'] = 1
df.loc[(df['Fare']>14.454) & (df['Fare']<=31.0),'Fare_Band'] = 2
df.loc[(df['Fare']>31.0) & (df['Fare']<=513),'Fare_Band'] = 3
```

```python
df.head()
```

```python
sns.factorplot('Fare_Band','Survived',hue='Sex', data=df)
```

- Irrespective of the Sex, the survival chances increases as the fare increases.
