```python
# create labels for the plot
x = np.arange(80,130,5)
y = np.arange(240,340,10)

plt.plot(x,y)

plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnt")

# title for the chart
plt.title("Sports Watch Data")
```

```python
# setting the font properties
x = np.arange(80,130,5)
y = np.arange(240,340,10)

font1 = {'family':'serif','color':'black','size':20} # 1st font property for labels
font2 = {'family':'serif','color':'darkred','size':25} # 2nd font property for Title

plt.plot(x,y)


plt.xlabel("Average Pulse", fontdict = font1)
plt.ylabel("Calorie Burnt", fontdict =font1)

# title for the chart
plt.title("Sports Watch Data", fontdict = font2)
```

```python
# changing the position of the title
# setting the font properties
x = np.arange(80,130,5)
y = np.arange(240,340,10)

font1 = {'family':'serif','color':'black','size':20} # 1st font property for labels
font2 = {'family':'serif','color':'darkred','size':25} # 2nd font property for Title

plt.plot(x,y)


plt.xlabel("Average Pulse", fontdict = font1)
plt.ylabel("Calorie Burnt", fontdict =font1)

# title for the chart
plt.title("Sports Watch Data", fontdict = font2, loc='left')
```

```python
# adding the gridlines to the chart

# changing the position of the title
# setting the font properties
x = np.arange(80,130,5)
y = np.arange(240,340,10)

font1 = {'family':'serif','color':'black','size':20} # 1st font property for labels
font2 = {'family':'serif','color':'darkred','size':25} # 2nd font property for Title

plt.plot(x,y, marker='o')


plt.xlabel("Average Pulse", fontdict = font1)
plt.ylabel("Calorie Burnt", fontdict =font1)

# title for the chart
plt.title("Sports Watch Data", fontdict = font2, loc='left')

# add the grid
plt.grid()

plt.show()
```

# Subplots

## a. Display Mutiple Plots

```python
# Draw 2 plots

# plot-1 parameters
x1 = np.array([0,1,2,3])
y1 = np.array([3,8,1,10])

# plot-2 parameters
x2 = np.array([0,1,2,3])
y2 = np.array([10,20,30,40])

# plotting subplot-1
plt.subplot(1,2,1) # the figure has 1 row, 2 columns and this is the first plot
plt.plot(x1,y1)

# plotting subplot-2
plt.subplot(1,2,2)
plt.plot(x2,y2)

plt.show()
```

```python
# Draw 2 plots

# plot-1 parameters
x1 = np.array([0,1,2,3])
y1 = np.array([3,8,1,10])

# plot-2 parameters
x2 = np.array([0,1,2,3])
y2 = np.array([10,20,30,40])

# plotting subplot-1
plt.subplot(2,1,1) # the figure has 2 row, 1 columns and this is the first plot
plt.plot(x1,y1)

# plotting subplot-2
plt.subplot(2,1,2)
plt.plot(x2,y2)

plt.show()
```

```python
# Draw 6-plots
# plot-1 parameters
x1 = np.array([0,1,2,3])
y1 = np.array([3,8,1,10])

# plot-2 parameters
x2 = np.array([0,1,2,3])
y2 = np.array([10,20,30,40])

# plot-3 parameters
x3 = np.array([0,1,2,3])
y3 = np.array([3,8,1,10])

# plot-4 parameters
x4 = np.array([0,1,2,3])
y4 = np.array([10,20,30,40])

# plot-5 parameters
x5 = np.array([0,1,2,3])
y5 = np.array([3,8,1,10])

# plot-6 parameters
x6 = np.array([0,1,2,3])
y6 = np.array([10,20,30,40])

plt.subplot(2,3,1)
plt.plot(x1,y1)

plt.subplot(2,3,2)
plt.plot(x2,y2)

plt.subplot(2,3,3)
plt.plot(x3,y3)

plt.subplot(2,3,4)
plt.plot(x4,y4)

plt.subplot(2,3,5)
plt.plot(x5,y5)

plt.subplot(2,3,6)
plt.plot(x6,y6)

plt.show()

```

```python
# Draw 6-plots : correcting the distance between the plots
# plot-1 parameters
x1 = np.array([0,1,2,3])
y1 = np.array([3,8,1,10])

# plot-2 parameters
x2 = np.array([0,1,2,3])
y2 = np.array([10,20,30,40])

# plot-3 parameters
x3 = np.array([0,1,2,3])
y3 = np.array([3,8,1,10])

# plot-4 parameters
x4 = np.array([0,1,2,3])
y4 = np.array([10,20,30,40])

# plot-5 parameters
x5 = np.array([0,1,2,3])
y5 = np.array([3,8,1,10])

# plot-6 parameters
x6 = np.array([0,1,2,3])
y6 = np.array([10,20,30,40])

plt.subplot(2,3,1)
plt.plot(x1,y1)
plt.title('x1 vs y1')

plt.subplot(2,3,2)
plt.plot(x2,y2)
plt.title('x2 vs y2')

plt.subplot(2,3,3)
plt.plot(x3,y3)
plt.title('x3 vs y3')

plt.subplot(2,3,4)
plt.plot(x4,y4)
plt.title('x4 vs y4')

plt.subplot(2,3,5)
plt.plot(x5,y5)
plt.title('x5 vs y5')

plt.subplot(2,3,6)
plt.plot(x6,y6)
plt.title('x6 vs y6')

plt.suptitle("Sales Data")  #super title

plt.tight_layout()

```

```python
# Sales data
x = np.array([5,7,8,7,2,17,18,25,80])
y = np.array([99,89,86,87,79,65,50,77,84])
sizes=np.array([20,50,1000,860,70,92,60,40,100])
colors = np.array([10,20,30,40,50,60,70,80,90])
plt.scatter(x,y,c = colors,  s=sizes, alpha =.4) #inbuilt color pallete
plt.colorbar()
```

```python
x = np.random.randint(100, size=(100))
y = np.random.randint(100,size=(100))
colors = np.random.randint(100,size=(100))
sizes = 10 * np.random.randint(100,size=(100))

plt.scatter(x,y,c =colors, s=sizes, alpha =0.5, cmap = 'nipy_spectral')

plt.colorbar()

plt.show()
```

# Matplotlib Bars

```python
# vertical bar plots
x= np.array(['A','B','C','D','E'])
y = np.array([4,9,2,11,16])

plt.bar(x,y)
plt.show()
```

```python
# Horizontal bar plots

x= np.array(['A','B','C','D','E'])
y = np.array([4,9,2,11,16])

plt.barh(x,y) # horizontal bar chart
plt.show()
```

```python
# Horizontal bar plots, coloring bars

x= np.array(['A','B','C','D','E'])
y = np.array([4,9,2,11,16])

plt.barh(x,y, color='red') # horizontal bar chart
plt.show()
```

```python
# Bar width : it will work only in vertical charts

x= np.array(['A','B','C','D','E'])
y = np.array([4,9,2,11,16])

plt.bar(x,y, color='red',width=0.2) # horizontal bar chart
plt.show()
```

```python
# bar height works in horizontal charts

x= np.array(['A','B','C','D','E'])
y = np.array([4,9,2,11,16])

plt.barh(x,y, color='red', height = 0.1) # horizontal bar chart
plt.show()
```

# Histogram

```python
x = np.random.normal(170,10,250)
plt.hist(x)
plt.show()
```

```python
# create a histogram with 15 buckets/bins
x = np.random.normal(170,10,250)
plt.hist(x,bins=15)
plt.show()
```

```python
# create a histogram with density
x = np.random.normal(170,10,250)
plt.hist(x,bins=15, density=True)
plt.show()

# density = counts/(sum(counts) * np.diff(bins))
```

```python
# Plotting two histograms together : side by side plots
y = np.random.normal(180,20,150)
plt.hist([x,y],stacked=False)
plt.show()
```

```python
# Plotting two histograms together : stacked plot : one plot on top of other plot
y = np.random.normal(180,20,150)
plt.hist([x,y],stacked=True)
plt.show()
```

# Pie Charts

```python
y = np.array([35, 25, 25, 15])
plt.pie(y)
plt.show()
```

```python
# insert labels to this pie chart
y = np.array([35, 25, 25, 15])
my_labels = ['English','Maths','Science','Hindi']
plt.pie(y, labels= my_labels)
plt.show()
```

```python
# putting the pie chart at 90 degree angle

# insert labels to this pie chart
y = np.array([35, 25, 25, 15])
my_labels = ['English','Maths','Science','Hindi']
plt.pie(y, labels= my_labels, startangle=90)
plt.show()
```

```python
# exploding the piece

# insert labels to this pie chart
y = np.array([35, 25, 25, 15])
my_labels = ['English','Maths','Science','Hindi']
my_explode = [0,0,.1,0]
plt.pie(y, labels= my_labels, startangle=90, explode=my_explode)
plt.show()
```

```python
# Beautify pie chart : add shadow and colors

# exploding the piece

# insert labels to this pie chart
y = np.array([35, 25, 25, 15])
my_labels = ['English','Maths','Science','Hindi']
my_explode = [0,0,.1,0]
colors =['orange','black','magenta','cyan']
plt.pie(y, labels= my_labels, startangle=90, explode=my_explode, shadow=True, colors=colors)
plt.show()
```

```python
# Adding the legend
# Beautify pie chart : add shadow and colors

# exploding the piece

# insert labels to this pie chart
y = np.array([35, 25, 25, 15])
# my_labels = ['English','Maths','Science','Hindi']
my_explode = [0,0,.1,0]
colors =['orange','black','magenta','cyan']
plt.pie(y, labels= my_labels, startangle=90, explode=my_explode, shadow=True, colors=colors)
plt.legend(loc='best')
plt.show()
```

```python
# add header to the legend
# Adding the legend
# Beautify pie chart : add shadow and colors

# exploding the piece

# insert labels to this pie chart
y = np.array([35, 25, 25, 15])
my_labels = ['English','Maths','Science','Hindi']
my_explode = [0,0,.1,0]
colors =['orange','yellow','magenta','cyan']
plt.pie(y, labels= my_labels, startangle=90, explode=my_explode, shadow=True, colors=colors, autopct ='%2.1f%%')
plt.legend(title ='Marks', loc='best')
plt.show()
```

```python
# load the dataset
df = sns.load_dataset('iris')

# print few rows 
df.head()
```

```python
df['species'].unique()
```

```python
# using seaborn let us plot the lineplot
sns.lineplot(x ='sepal_length', y='sepal_width', data=df)
```

```python
# using seaborn let us plot the lineplot
sns.lineplot(x ='sepal_length', y='sepal_width', data=df)
plt.title("IRIS DATASET")
plt.show()
```

```python
# changing the x and y limits

sns.lineplot(x ='sepal_length', y='sepal_width', data=df)
plt.xlim(5) # start from 5
plt.ylim(2) #start from 2
plt.title("IRIS DATASET")
plt.show()
```

```python
#change the background style
sns.lineplot(x ='sepal_length', y='sepal_width', data=df)

#change the style
sns.set_style("whitegrid")
plt.xlim(5) # start from 5
plt.ylim(2) #start from 2
plt.title("IRIS DATASET")
plt.show()
```

```python
# change the figure size

plt.figure(figsize=(4,4))
sns.lineplot(x ='sepal_length', y='sepal_width', data=df)

#change the style
sns.set_style("whitegrid")
plt.xlim(5) # start from 5
plt.ylim(2) #start from 2
plt.title("IRIS DATASET")
plt.show()
```

```python
# change the scale of the plots

plt.figure(figsize=(8,4))
sns.lineplot(x ='sepal_length', y='sepal_width', data=df)


plt.xlim(5) # start from 5
plt.ylim(2) #start from 2

#changing the scale of the plot
sns.set_context("paper", font_scale=2)
plt.title("IRIS DATASET")
plt.show()
```
