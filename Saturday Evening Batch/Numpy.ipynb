# Vectorization Functions

```python
def my_func(a,b):
    '''Return a-b if a>b otherwise it will return b-a'''
    if a>b:
        return a-b
    else:
        return b-a
```

```python
vfunc = np.vectorize(my_func)
```

```python
a = np.arange(-9,9,2)
b = np.arange(-9,9,2).reshape((3,3))
print(a,"\n")
print(b)
```

```python
vfunc(a,10)
```

```python
vfunc(b,5)
```

# Joining Array

```python
base_array = np.arange(1,10).reshape((3,3))
join_array = np.arange(1,4).reshape((1,3))
```

```python
print(base_array)
```

```python
print(join_array)
```

```python
np.concatenate((base_array,join_array),axis=0)
```

```python
join_array.T
```

```python
np.concatenate((base_array,join_array.T),axis=1)
```

```python
np.vstack((base_array,join_array))
```

```python
np.hstack((base_array,join_array.T))
```

```python
np.r_[base_array,join_array]
```

```python
np.c_[base_array,join_array.T]
```

```python
# concatenate : axis=0 -> tuples, vstack: tuple, r_ :[]
#cancatenate :axis=1-> tuples, hstack :tuples, c_:[]
```
