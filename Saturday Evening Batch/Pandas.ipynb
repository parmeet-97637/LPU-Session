# Why Pandas? or What is Pandas?

Pandas can handle the different types of data
- Tabular Data : Excel :Rows &Cols in similar way tabular data is k/as combination of rows and cols.
- It can handle unstructured data.. 
- ordered and unordered data.
- homogeneous data

#### Pandas can handle 2 types of data structures
- Series (1-dimensional)
- DataFrames (2-D)

# What Pandas can do.?

- missing data
- Insert/Delete operations on DataFrame
- We can align the data (Data Alignment/reindexing)
- We can group the data
- we can convert data into different forms/formats
- We can performing joining and merging operations
- reshaping
- pivoting
- we can load flat files (Text files)
- We can time series data

# How to Filter Rows

```python
train.head()
```

```python
# I want to select the passengers older than 35 years
condition = train['Age'] >35
```

```python
condition
```

```python
# fetch the filtered rows
train[condition] # train[train['Age'] >35]
```

```python
# I want to select the survived male passenegers 
#cols : survived,sex 
condition1 = train['Sex'] == 'male' #selected male passengers
condition2 = train['Survived'] ==1 # selected survived people

condition = (condition1 & condition2)
```

```python
# filter the rows
# train[condition] 
train[(train['Sex'] == 'male') & (train['Survived'] ==1)]
```

```python
# I want to select the survived male passenegers and all the people who boarded ship from port "S"

#cols = survived, sex, embarked 
condition1 = train['Embarked'] =='S'
condition2 = train['Sex'] == 'male' #selected male passengers
condition3 = train['Survived'] ==1 # selected survived people

condition = (condition1 | (condition2 & condition3))
```

```python
train[condition]
```
